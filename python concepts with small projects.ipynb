```python
#1.Student Marks Analyzer 

marks={
    "Alice": [75, 80, 90], 
    "Bob": [60, 70, 65], 
    "Charlie": [90, 95, 88] 
}
print("Give average : ")
for name in marks:
    average=sum(marks[name])/3
    print(f"{name} having average :{avg:.1f}")

max_total = 0 
topper = "" 
for name in marks: 
    total = sum(marks[name]) 
    if total > max_total: 
        max_total = total 
        topper = name
        
print(f"\n topper:{topper } with {max_total} marks.")

print("Subject-wise averages: ")
subjects=["Math", "Science","English"]

for i in range(2):
    total=0
    for name in marks :
        total += marks[name][i]
    avg = total/len(marks)
    print(f"{subjects[i]}:{avg:.1f}")
        
```

```python
# contacr_book

contacts = {}

def add_contact():
    name = input("Enter name: ").lower()
    phone = input("Enter phone: ")
    contacts[name] = phone
    print("Contact added!")

def view_contacts():
    if not contacts:
        print("No contacts found.")
    else:
        for name, phone in contacts.items():
            print(f"{name} - {phone}")

def search_contact():
    name = input("Enter name to search: ").lower()
    if name in contacts:
        print(f"Found: {name} - {contacts[name]}")
    else:
        print("Contact not found.")

def update_contact():
    name = input("Enter name to update: ").lower()
    if name in contacts:
        contacts[name] = input("Enter new phone: ")
        print("Updated successfully!")
    else:
        print("Contact not found.")

def delete_contact():
    name = input("Enter name to delete: ").lower()
    if name in contacts:
        del contacts[name]
        print("Deleted successfully!")
    else:
        print("Contact not found.")

while True:
    print("\n Contact Book Menu ")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        update_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
        print("Goodbye!")
        break
    else:
        print("Invalid choice!")
```

```python
#3. Banking system

def create_account(data):
    while True:
        try:
            count = int(input("Enter accounts you want to add: "))
            break
        except ValueError:
            print("Enter valid input")
            continue
    for i in range(count):
        name = input("Enter customer's name: ").strip().lower()
        if not name:
            print("Name cannot be empty!")
            continue
        if name in data:
            print('Account already exists. Try updating instead.')
            continue
        else:
            amount = get_valid_amount("Enter amount: ")
            data[name] = amount
    print("Account created successfully!")

def check_balance(data):
    name = input("Enter account holder name to check the balance: ").strip().lower()
    if name not in data:
        print("Account not found")
        return

    print(f"{name}'s balance: {data[name]}")

def deposit(data):
    name = input("Enter account holder name: ").strip().lower()
    if name not in data:
        print("Account not found")
        return

    amount = get_valid_amount("Enter amount to deposit: ")
    data[name] += amount
    print("Deposit successful!!")

def withdraw(data):
    name = input("Enter account holder name: ").strip().lower()
    if name not in data:
        print("Account not found")
        return

    amount = get_valid_amount("Enter amount to withdraw: ")
    if data[name] >= amount:
        data[name] -= amount
        print("Withdrawal successful  !!")
    else:
        print("Insufficient balance")

def get_valid_amount(prompt):
    while True:
        try:
            amount = int(input(prompt))
            if amount < = 0:
                print("Amount must be positive")
                continue
            return amount
        except ValueError:
            print("Enter valid number")

account_data = {}

while True:
    print("\n Welcome to MyBank ")
    print("1.Create an account\n2.Check balance\n3.Deposit money\n4.Withdraw money\n5.Exit\n")
    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print('Please enter a valid number from the menu.')
        continue

    if choice == 1:
        create_account(account_data)
    elif choice == 2:
        check_balance(account_data)
    elif choice == 3:
        deposit(account_data)
    elif choice == 4:
        withdraw(account_data)
    elif choice == 5:
        print("Thank you for using MyBank!")
        break
    else:
        print("Invalid.")
        
```
